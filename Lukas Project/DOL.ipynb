```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup

# Function to clean text by replacing multiple spaces with a single space and removing excessive newlines
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces/newlines with a single space
    return text.strip()

# CSV formation
with open('DOLpressreleases3.csv', 'w', newline='', encoding='utf-8') as f:
    w = csv.DictWriter(f, fieldnames=["Date", "URL", "Title", "Text"])
    w.writeheader()

    driver_path = Service('/Users/ea025/Desktop/Python Camp - Alexander/PythonCamp2024/Lukas Project/chromedriver')
    driver = webdriver.Chrome(service=driver_path)
    
    # DOL press releases *START*
    for page_num in range(401, 470): 
        url = f'https://www.dol.gov/newsroom/releases?page={page_num}'
        driver.get(url)
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)  
        
        # Parse - BeautifulSoup
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
        # Find and process each press release on the page
        releases = soup.find_all('div', class_='image-left-teaser')

        for release in releases:  # Loop through each press release item
            try:
                fac = {}  # Empty dict to store the press release data
                
                # Extract Date
                date_elem = release.find('p', class_='dol-date-text')
                if date_elem:
                    fac['Date'] = date_elem.text.strip()
                else:
                    fac['Date'] = ''

                # Extract URL
                url_elem = release.find('a')
                if url_elem and 'href' in url_elem.attrs:
                    fac['URL'] = f"https://www.dol.gov{url_elem['href'].strip()}"
                else:
                    fac['URL'] = ''
                
                # Extract Title
                title_elem = release.find('h3')
                fac['Title'] = title_elem.text.strip() if title_elem else ''

                # Only proceed if URL is valid
                if fac['URL']:
                    # Open the interior page using Selenium
                    driver.get(fac['URL'])
                    time.sleep(3)  # Allow time for page load
                    
                    # Scrape the interior page
                    interior_html = driver.page_source
                    interior_soup = BeautifulSoup(interior_html, 'html.parser')

                    # Extract Text from the <div> with class 'field--name-field-press-body'
                    body_elem = interior_soup.find('div', class_='field--name-field-press-body')
                    if body_elem:
                        raw_text = body_elem.get_text(separator='\n').strip()
                        fac['Text'] = clean_text(raw_text)  # Clean the extracted text
                    else:
                        fac['Text'] = ''

                    # Write the data to CSV
                    w.writerow(fac)
        
            except Exception as e:
                print(f"Error processing release: {e}")
                continue  # Skip to the next item in case of an error

    # Close the driver after all pages have been processed
    driver.quit()

print("All done!")
```
