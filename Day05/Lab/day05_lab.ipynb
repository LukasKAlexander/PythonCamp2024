##### Instructions: 

Write your own Python script to answer the following questions: 
1. Which of these embassies is closest to the White House in meters? What is the address? 
2. If I wanted to hold a morning meeting there, which cafe would you suggest (best rating and closest)?
3. If I wanted to hold an upscale evening meeting there, which fancy bar would you suggest? 

Hint: 
- You will need to enable the `Google Places API`.
- You may find this page useful to learn about different finding nearby places https://www.geeksforgeeks.org/python-fetch-nearest-hospital-locations-using-googlemaps-api/

```python
import importlib
import googlemaps

imported_items = importlib.import_module('start_google')
gmaps = imported_items.client

whitehouse = (38.8976763, -77.0365298)

embassies = [
    (38.917228, -77.0522365), 
    (38.9076502, -77.0370427), 
    (38.916944, -77.048739)
]

distance1 = gmaps.distance_matrix(whitehouse, embassies[0])
distance2 = gmaps.distance_matrix(whitehouse, embassies[1])
distance3 = gmaps.distance_matrix(whitehouse, embassies[2])

print(f"Distance to Embassy 1: {distance1['rows'][0]['elements'][0]['distance']['text']}")
print(f"Distance to Embassy 2: {distance2['rows'][0]['elements'][0]['distance']['text']}")
print(f"Distance to Embassy 3: {distance3['rows'][0]['elements'][0]['distance']['text']}")

#Embassy 2 is closest

#part 2

whitehouse = (38.8976763, -77.0365298)

places_result = gmaps.places_nearby(
    location=whitehouse, 
    radius=500, 
    type='cafe'
)

best_cafe = None
highest_rating = 0

for place in places_result['results']:
    if 'rating' in place and place['rating'] > highest_rating:
        highest_rating = place['rating']
        best_cafe = place

if best_cafe:
    cafe_name = best_cafe['name']
    cafe_address = best_cafe['vicinity']
    cafe_rating = best_cafe['rating']
    
    distance = gmaps.distance_matrix(whitehouse, best_cafe['geometry']['location'])
    distance_text = distance['rows'][0]['elements'][0]['distance']['text']
    
    print(f"Best Cafe: {cafe_name}")
    print(f"Address: {cafe_address}")
    print(f"Rating: {cafe_rating}")
    print(f"Distance from White House: {distance_text}")
else:
    print("No suitable cafes found nearby.")

#part 3

whitehouse = (38.8976763, -77.0365298)

places_result = gmaps.places_nearby(
    location=whitehouse, 
    radius=500, 
    type='bar'
)

best_bar = None
highest_rating = 0

for place in places_result['results']:
    if 'rating' in place and place['rating'] > highest_rating:
        highest_rating = place['rating']
        best_bar = place

if best_bar:
    bar_name = best_bar['name']
    bar_address = best_bar['vicinity']
    bar_rating = best_bar['rating']
    
    distance = gmaps.distance_matrix(whitehouse, best_bar['geometry']['location'])
    distance_text2 = distance['rows'][0]['elements'][0]['distance']['text']
    
    print(f"Best Bar: {bar_name}")
    print(f"Address: {bar_address}")
    print(f"Rating: {bar_rating}")
    print(f"Distance from White House: {distance_text2}")
else:
    print("No suitable bars found nearby.")
```
